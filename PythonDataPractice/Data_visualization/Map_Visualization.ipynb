```python
import pandas as pd
```

```python
file_path = './data/서울시 대피소 방재시설 현황 (좌표계_ WGS1984).csv'
raw = pd.read_csv(file_path, encoding = 'cp949')
```

```python
raw.head()
```

```python
import folium
from folium.plugins import MiniMap
from folium.plugins import MarkerCluster
```

```python
# 지도 생성 (중심좌표 : 서울역)
m = folium.Map(location = [37.5536067,126.9674308],
              zoom_start = 11)
# 클러스터 마커 추가
markercluster = MarkerCluster().add_to(m)

# 미니맵 추가
miniM = MiniMap()
miniM.add_to(m)

for i in range(len(raw)):
    lat = raw.loc[i,'위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    max_num = raw.loc[i, '최대수용인원']
    now_num = raw.loc[i, '현재수용인원']
    folium.Marker([lat, long],
                 tooltip = name,
                 popup = f'최대수용인원:{max_num}, 현재인원:{now_num}').add_to(markercluster)
    
m
```

```python
m.save('./data/MapOfShelter.html')
```

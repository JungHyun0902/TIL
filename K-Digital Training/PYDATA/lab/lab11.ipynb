```python
#1
import pandas as pd

df1 = pd.read_csv("./data/score.csv")
print('컬럼명 : ', df1.columns)
print('인덱스 : ', df1.index)
df1['total'] = df1.kor + df1.eng + df1.mat
df1['avg'] = round(df1.total/3, 2)
display(df1)
```

```python
#2
df2 = pd.read_csv("./data/score_noheader.csv", header = None)
display(df2)
print('컬럼명 : ', df2.columns)
print('인덱스 : ', df2.index)
print('컬럼명 추가 --------------')
df2.columns = ['name', 'kor', 'eng', 'mat']
print('컬럼명 : ', df2.columns)
display(df2)
print('total과 avg열 추가 --------')
df2['total'] = df2.kor + df2.eng + df2.mat 
# pd.options.display.float_format = '{:.2f}'.format
df2['avg'] = round(df2.total/3, 2)
display(df2)
```

```python
#3
df3 = pd.read_csv("./data/score_header.csv", header = 3)
display(df3)
print('컬럼명 : ', df3.columns)
print('인덱스 : ', df3.index)
df3['total'] = df3.kor + df3.eng + df3.mat 
pd.options.display.float_format = '{:.2f}'.format
df3['avg'] = df3.total/3
display(df3)
df3.to_csv("output/score_result.csv", header = False, index = False)
```

```python
#4
df4 = pd.read_excel("./data/score.xlsx")
display(df4)
df4['total'] = df4.kor + df4.eng + df4.mat 
pd.options.display.float_format = '{:.2f}'.format
df4['avg'] = df4.total/3
ndf4 = df4.sort_values(by='total', ascending = False)
display(ndf4)

ndf4.to_excel("output/score_result1.xlsx")
ndf4.to_excel("output/score_result2.xlsx", index = False)
```
